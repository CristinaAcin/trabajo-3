```python
#Es por ello que para trabajar mejor directamente eliminamos las filas que contengan algún NA
df=trabajo3.dropna()
df.head()
```

```python
df.count()
```

**ANÁLISIS EXPLORATORIO**

En primer lugar vamos a comprobar si los clientes están repetidos en el dataset (CUST_ID) o sin embargo cada fila representa a una persona. 

```python
df["CUST_ID"].unique()
```

```python
df.groupby('CUST_ID').mean().count()
```

Comprobamos que cada fila corresponde a un único cliente, no están repetidos. En total hay 8636 clientes.

```python
#Vemos la distribución de la variable BALANCE
sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=df["BALANCE"])
```

```python
df.columns
```

```python
df.head(5)
```

```python
fig, ax = plt.subplots(figsize=(7,4), dpi=300)

sns.scatterplot(data=df, x='BALANCE', y='PURCHASES', hue='CREDIT_LIMIT', palette='turbo', ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(7,4), dpi=300)

sns.histplot(data=df, x='BALANCE', hue='TENURE',  palette='turbo', ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(7,4), dpi=300)

sns.histplot(data=df[df.TENURE < 12], x='BALANCE', hue='TENURE',  palette='turbo', ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(7,4), dpi=300)


sns.barplot(data=df, x='TENURE', y='CREDIT_LIMIT',label='Credit Limit', color='lightblue')
sns.barplot(data=df, x='TENURE', y='BALANCE',label='Balance', color='b')

ax.legend()
```
